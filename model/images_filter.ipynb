```python
import os
from PIL import Image

metacarpal_img_list = []

for index, image_name in enumerate(y_train):
    image_path = os.path.join('./cropped_images_png/metacarpal/training/',str(index)+'.png')
    img = Image.open(image_path)
    img_array = np.array(img)
    metacarpal_img_list.append(img_array)

carpal_img_list = []

for index, image_name in enumerate(y_train):
    image_path = os.path.join('./cropped_images_png/carpal/training/',str(index)+'.png')
    img = Image.open(image_path)
    img_array = np.array(img)
    carpal_img_list.append(img_array)
```

```python
import os
X_val_carpal_filter = []
X_val_metacarpal_filter = []
y_val_filter = []
gender_val_filter = []

carpal_img_num_list = [int(name.strip('.jpg')) for name in os.listdir('./cropped_images_drive/carpal/validation/')]
metacarpal_img_num_list = [int(name.strip('.jpg')) for name in os.listdir('./cropped_images_drive/metacarpal/validation/')]

carpal_img_num_list.sort()
metacarpal_img_num_list.sort()

for index in metacarpal_img_num_list:
    if index in carpal_img_num_list:
        X_val_carpal_filter.append(carpal_img_list[index])
        X_val_metacarpal_filter.append(metacarpal_img_list[index])

        y_val_filter.append(y_val[index])
        gender_val_filter.append(gender_val[index])
```
